```python
# Не нужные данные
columns_to_drop = ['id', 'spkid', 'name', 'prefix', 'orbit_id', 'epoch_mjd', 'equinox',
                   'sigma_e', 'sigma_a', 'sigma_q', 'sigma_i', 'sigma_om', 'sigma_w',
                   'sigma_ma', 'sigma_ad', 'sigma_n', 'sigma_tp', 'sigma_per', 'rms']


df = df.drop(columns=columns_to_drop)
df.head()
```

```python
df.shape
```

```python
df = df.drop_duplicates()
```

```python
print(df.isnull().sum())
```

```python
# Замена NULL-значений на моду в указанных категориальных столбцах
df['neo'].fillna(df['neo'].mode().iloc[0], inplace=True)
df['pha'].fillna(df['pha'].mode().iloc[0], inplace=True)

# Замена NULL-значений на 0 в указанных столбцах
columns_to_replace_zero = ['diameter', 'albedo', 'diameter_sigma']
df[columns_to_replace_zero] = df[columns_to_replace_zero].fillna(0)

# Замена NULL-значений на среднее значение
df['H'].fillna(df['H'].mean(), inplace=True)
df['ma'].fillna(df['ma'].mean(), inplace=True)
df['ad'].fillna(df['ad'].mean(), inplace=True)
df['per'].fillna(df['per'].mean(), inplace=True)
df['per_y'].fillna(df['per_y'].mean(), inplace=True)

df['moid'].fillna(df['moid'].mean(), inplace=True)
df['moid_ld'].fillna(df['moid_ld'].mean(), inplace=True)
```

# 4 (Optional) Make an Hypothesys on your dataset and try to confirm or reject it

```python
omb_data = df[df['class'].str.upper() == 'OMB']['diameter'].dropna()
mba_data = df[df['class'].str.upper() == 'MBA']['diameter'].dropna()
```

```python
# Расчет стандартного отклонения для каждой выборки
std_omb = omb_data.std()
std_mba = mba_data.std()
```

```python
# Стандартное отклонение
print(f'Стандартное отклонение OMB: {std_omb}')
print(f'Стандартное отклонение MBA: {std_mba}')
```

```python
# Выполняем T-тест
t_stat, p_value = ttest_ind(omb_data, mba_data, equal_var=False)
```

```python
# Выводим результаты
print(f'T-статистика: {t_stat}')
print(f'P-значение: {p_value}')
```

```python
    alpha = 0.05
    if p_value < alpha:
        print('Отвергаем нулевую гипотезу. Существует статистически значимая разница в средних значениях диаметров.')
    else:
        print('Не удалось отвергнуть нулевую гипотезу. Нет статистически значимой разницы в средних значениях диаметров.')
```

Гипотеза: Средние диаметры астероидов различаются в зависимости от их классификации.

# 5 Make an Regression, choose proper model to predict insight depending on your dataset

```python
sns.countplot(x='neo', data=df)
```

```python
df['neo'].replace({'Y': 1, 'N': 0}, inplace=True)
```

```python
df['neo'].sum()
```

```python
# Определение признаков и целевой переменной
X = df[["diameter", "diameter_sigma", "a", "i", "ad",  "moid_ld", "moid",  "moid_ld"]]
print(X)
Y = df["neo"]
print(Y)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X, Y)
status = model.predict(X)
accuracy = accuracy_score(Y, status)
slope = model.coef_
```

```python
print("Coefficients:", slope)
```

```python
# Вывод результатов
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
conf_matrix = confusion_matrix(Y, status)
print("Confusion Matrix:")
print(conf_matrix)
```

Результат говорит, что модель правильно предсказала 935210+22482, и неправильно 413+419.

```python
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Не близко', 'Близко'], yticklabels=['Не близко', 'Близко'])
plt.xlabel('Предсказано')
plt.ylabel('Фактически')
plt.title('Confusion Matrix')
plt.show()
```
